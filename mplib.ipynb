```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

x = np.random.randint(0,55,55)
y = np.random.randint(0,55,55)

# Plotting the data
plt.scatter(x, y,marker="d")
plt.xlabel("bruh")
plt.ylabel("lookin like the battle pass")
plt.title("guy from fortnite")
# Displaying the plot
plt.show()
```

```python
categories = ['number of values', 'value of numbers', 'Green', 'Purple']
values = np.random.randint(0,55,4)

# Creating the bar chart
plt.bar(categories, values)

# Adding title and labels
plt.title('Simple Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')

# Displaying the plot
plt.show()
```

```python
# Data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Plotting with a specific color and marker
plt.plot(x, y, marker='d', color='green')  # 's' stands for square marker

# Adding title and labels
plt.title('Colorful Plot with diamond Markers')
plt.xlabel('Number go Left')
plt.ylabel('Number go Up')

# Displaying the plot
plt.show()
```
